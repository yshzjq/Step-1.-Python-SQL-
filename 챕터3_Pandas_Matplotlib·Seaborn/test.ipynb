```python
import matplotlib.pyplot as plt

# Windows: 맑은 고딕 / Colab·Linux: 나눔고딕
import platform
font = 'Malgun Gothic' if platform.system() == 'Windows' else 'NanumGothic'
plt.rc('font', family=font)
plt.rc('axes', unicode_minus=False)  # 마이너스 기호 깨짐 방지

# 테스트
plt.figure(figsize=(5, 2))
plt.text(0.5, 0.5, '한글 테스트 성공!', fontsize=20, ha='center')
plt.axis('off')
plt.show()
```
