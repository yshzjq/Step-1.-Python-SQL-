```python
import pandas as pd

# 세 가지 데이터
A = pd.Series([5, 5, 5, 5, 5])      # 전부 같음
B = pd.Series([3, 4, 5, 6, 7])      # 조금 퍼짐
C = pd.Series([0, 0, 5, 10, 10])    # 많이 퍼짐

print(f"A: 평균={A.mean()}, 표준편차={A.std():.2f}")
print(f"B: 평균={B.mean()}, 표준편차={B.std():.2f}")
print(f"C: 평균={C.mean()}, 표준편차={C.std():.2f}")
```
