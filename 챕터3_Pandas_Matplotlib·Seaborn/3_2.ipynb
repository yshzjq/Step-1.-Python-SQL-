```python
# 결측치 개수 세기
with open('seoul-rainfall-data.csv', 'r', encoding='cp949') as f:
    lines = f.readlines()

# 첫 줄은 헤더(컬럼명)이므로 제외
data_lines = lines[1:]

total_count = 0      # 전체 데이터 수
missing_count = 0    # 결측치 수

for line in data_lines:
    line = line.strip()  # 줄바꿈 제거
    if line:  # 빈 줄이 아니면
        parts = line.split(',')  # 쉼표로 분리
        rainfall = parts[2]      # 세 번째 값 = 강수량

        total_count += 1
        if rainfall == '':       # 비어있으면 결측치
            missing_count += 1

print(f"전체 데이터: {total_count}개")
print(f"결측치: {missing_count}개")
print(f"결측치 비율: {missing_count / total_count * 100:.1f}%")
```

```python
# 학습용 예시 데이터: 일주일 강수량
rainfall_data = ["0", "5.2", "", "12.3", "", "", "3.1"]

# 여기에 코드 작성

missing_count = 0 # 결측치 수

for data in rainfall_data :
  if data == '':
    missing_count+=1

print(f"전체 데이터: {len(rainfall_data)}개")
print(f"결측치: {missing_count}개")
print(f"결측치 비율: {missing_count / len(rainfall_data) * 100:.1f}%")
```

```python
rainfall_data = ["0", "5.2", "", "12.3", "", "", "3.1"]

new_rainfall_data = []

for data in rainfall_data :
  if data == "" :
    new_rainfall_data.append(0)
  else :
    new_rainfall_data.append(float(data))

print(f"원본: {rainfall_data}")
print(f"정제: {new_rainfall_data}")
print(f"합계: {sum(new_rainfall_data)}mm")
print(f"평균: {sum(new_rainfall_data) / len(new_rainfall_data):.1f}mm")
```
