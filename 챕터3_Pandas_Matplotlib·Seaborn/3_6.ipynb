```python
rainfall = pd.read_csv('seoul-rainfall-data.csv',encoding='cp949')
sales = pd.read_csv('cafe-sales-weekly.csv')

rainfall.head(5)
```

```python
sales.head(5)
```

```python
weekly_sales_sum = sales.groupby('주')['매출'].sum()
print(weekly_sales_sum.head(10))
```

```python
weekly_sales_sum = sales.groupby('카테고리')['매출'].agg(['sum', 'mean', 'count'])
print(weekly_sales_sum.head(10))
```

```python
weekly_sales_sum = sales.groupby(['매장','카테고리'])['매출'].agg(['sum', 'mean', 'count'])
print(weekly_sales_sum.head(10))
```

```python
rainfall['날짜'] = pd.to_datetime(rainfall['날짜'])

rainfall['주'] = 'W' + rainfall['날짜'].dt.isocalendar().week.astype(str).str.zfill(2)

print(rainfall['날짜'].head(15))

print(rainfall['날짜'].dt.isocalendar().week.head(15))


print(rainfall.head(10))
```
