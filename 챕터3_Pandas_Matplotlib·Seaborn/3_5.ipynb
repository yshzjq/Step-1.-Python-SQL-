```python
import pandas as pd
df = pd.read_csv('seoul-rainfall-data.csv', encoding='cp949')

print(df['강수량(mm)'].isnull().sum())
df['강수량(mm)'].fillna(0)
print(df['강수량(mm)'].isnull().sum())

df['정제된 데이터'] = df['강수량(mm)'].fillna(0)
print(df['정제된 데이터'].isnull().sum())
```

```python
df['정제된 데이터'] = df['강수량(mm)'].fillna('Unknown')
print(df['정제된 데이터'].isnull().sum())
```

```python
print(df['강수량(mm)'].sum()) # 총합
print(df['강수량(mm)'].isnull()) # 결측치 여부 확인
print(df['강수량(mm)'].isnull().sum()) # 결측치 몇개인지 True가 결측치
```

```python
df.isnull().sum()
```

```python
missing_count = df['강수량(mm)'].isnull().sum()
total_count = len(df)
missing_ratio = missing_count / total_count * 100
print(f"결측치 비율 : {missing_ratio:.2f}%")
```

```python
print(len(df))
df_cleaned = df.dropna() # 결측치 존재할 경우 다 삭제
print(len(df_cleaned)) #

df_cleaned = df.dropna(how='all')
print(len(df_cleaned)) #

df_cleaned = df.dropna(subset=['강수량(mm)'])
print(len(df_cleaned)) #
```

**장점**

구현이 간단함

결측치 패턴이 무작위일 때 효과적

분석 결과가 왜곡되지 않음


단점

데이터 솔실 발생

결측치 많으면 대부분 삭제됨

편향(bias)이 생길 수 있음



```python

# 평균(mean)으로 채우기
df['강수량(mm)'].fillna(df['강수량(mm)'].mean())

# 중앙값(median)으로 채우기 - 이상치에 덜 민감
df['강수량(mm)'].fillna(df['강수량(mm)'].median())

# 최빈값(mode)으로 채우기 - 범주형 데이터에 적합
df['강수량(mm)'].fillna(df['강수량(mm)'].mode()[0])
```
