```python
import pandas as pd
df = pd.read_csv('seoul-rainfall-data.csv', encoding='cp949')

print(df['강수량(mm)'].isnull().sum())
df['강수량(mm)'].fillna(0)
print(df['강수량(mm)'].isnull().sum())

df['정제된 데이터'] = df['강수량(mm)'].fillna(0)
print(df['정제된 데이터'].isnull().sum())
```

```python
df['정제된 데이터'] = df['강수량(mm)'].fillna('Unknown')
print(df['정제된 데이터'].isnull().sum())
```
