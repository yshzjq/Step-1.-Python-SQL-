```python
def clean_rainfall(value):
    """
    강수량 값을 정제하는 함수
    - 빈 문자열이면 0.0 반환
    - 그 외에는 숫자로 변환해서 반환
    """
    # 여기에 코드 작성
    if value.strip() == "" :

      return (float(0))
    else :
       return (float(value))



# 테스트
print(clean_rainfall(""))      # 예상: 0.0
print(clean_rainfall("5.2"))   # 예상: 5.2
print(clean_rainfall("0"))     # 예상: 0.0
```

```python
# 학습용 예시 데이터 (정제 완료된 상태)
rainfall_data = [0.0, 0.0, 0.6, 0.0, 0.0, 2.4, 0.0, 0.0, 1.6, 0.0]

# 여기에 코드 작성
result = 0
for data in rainfall_data :
  if data > float(0) :
    result+=1

print(result)
```

```python
# 학습용 예시 데이터
data = [
    ['2024-01-01', '108', 0.0],
    ['2024-01-02', '108', 0.6],
    ['2024-02-01', '108', 3.9],
    ['2024-02-02', '108', 0.2],
]

# 여기에 코드 작성
result = []
for i in range(13):
  result.append(float(0))

for d in data:
  month = int(d[0].split('-')[1])
  result[month] += d[2]


print(result)

for i in range(len(result) - 1):
  print(i+1," 월 강수량 : ", result[i+1])
```
