리스트로 총합과 평균을 계산하고
딕셔너리로 대표 행사 가격을 조회해서
한 문장으로 요약 리포트를 출력합니다

```python
price_by_event = {
    "[구로문화재단] 2026 구로문화재단 신년음악회": 10000,
    "[종로아이들극장] 미디어 상상 놀이극 [거인의 책상]": 30000,
    "서울일러스트레이션페어V.20": 15000,
    "[광진문화재단] 가족무용극 고블린파티② [루돌프]": 25000,
    "[강동문화재단] 목요예술무대Ⅲ [제비씨의 크리스마스]": 20000
}

ticket_prices = [10000, 30000, 15000, 25000, 20000]


count = len(ticket_prices)
total = sum(ticket_prices)
avg = total/count

price_by_event = {"[구로문화재단] 2026 구로문화재단 신년음악회" : 10000,
"[종로아이들극장] 미디어 상상 놀이극 [거인의 책상]" : 30000,
"서울일러스트레이션페어V.20" : 15000,
"[광진문화재단] 가족무용극 고블린파티② [루돌프]" : 25000,
"[강동문화재단] 목요예술무대Ⅲ [제비씨의 크리스마스]" : 20000
}

대표행사 = "서울일러스트레이션페어V.20"
대표가격 = price_by_event[대표행사]

report = (
    "서울시 문화행사 티켓 가격 요약 리포트: "
    + "총합 " + str(total) + "원, "
    + "평균 " + str(int(avg)) + "원, "
    + "대표 행사(" + 대표행사 + ") " + str(대표가격) + "원"
)

print(report)
```
