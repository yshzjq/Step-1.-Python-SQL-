```python
import random
import time
from datetime import datetime

start = time.time()

# 오늘 날짜 문자열 만들기
today_text = datetime.now().strftime("%Y/%m/%d")

# 랜덤 추천 1개 뽑기
picked = random.choice(events)

end = time.time()
elapsed = end - start

# 리포트 출력
print("서울시 문화 행사 일정 요약")
print("오늘 날짜:", today_text)
print()

print("이번 주 예정 행사들")
# 예제에서 사용한 코드 넣기

print("오늘의 추천 행사")
print("-", picked["date"], picked["name"])
print()

print("전체적으로", elapsed, "초 걸렸습니다.")
```
