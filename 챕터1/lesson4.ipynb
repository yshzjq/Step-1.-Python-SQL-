- `상황` 상품 가격이 문자열로 들어옴(쉼표 포함), 주문 건수는 숫자
    - total_sales_text = "123,000"
    - order_count = 12
- `실습`
    - 문자열에서 쉼표를 제외하고
    - 객단가를 계산하여
    - “총 매출 : XXXXX원. 주문건수: XX건. 객단가: XXXX원” 으로 출력

```python
total_sales_text = "123,000"
order_count = 12

total_sales_text = total_sales_text.replace(",", "")
result = int(total_sales_text) / order_count

print("총 매출 :", int(total_sales_text), "원. 주문건수:", order_count, "건. 객단가:", result, "원")
```
