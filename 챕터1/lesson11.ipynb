```python
target_date = "2025-11-15"

for row in sorted_events:
    if row["시작일"] >= target_date:
        print("조건 성립")
        print(row)
        break
    
```

```python
target_date = "2025-11-15"


left = 0;
right = len(sorted_events) - 1

result_index = None

while left <= right:
    mid = (left + right) // 2
    mid_date = sorted_events[mid]['시작일']

    if mid_date >= target_date :

        result_index = mid
        right = mid - 1

    else:
        left = mid + 1

if result_index is not None:
    print("조건에 맞는 행사를 찾았다!")
    print(sorted_events[result_index])
else:
    print("조건에 맞는 행사가 없다")


```
